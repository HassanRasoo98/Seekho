```python
def format_topic(topics):
  formatted_topics = []
  for topic in topics:
    res = ''
    for i in topic[1]:
      if ord(i) >= 97 and ord(i) <= 122 or ord(i) == 32:
        res += i
    formatted_topics.append(res)

  return formatted_topics
```
